<h1> Introduccion </h1>

El metodo de gradiente conjugado es la manera mas eficiente para poder resolver grandes sistemas de ecuaciones lineales como el caso de 

\begin{align}
    \label{eq_01}
    &Ax = b
\end{align}

Siendo $x$ el vector que no conocemos, $b$ un vector que si conocemos y $A$ puede ser una matriz cuadratica positiva definida o indefinida

Para la creacion de un algoritmo funcional capaz de resolver un ejercicio de estas caracteristicas utilizaremos python

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import itertools
```

La importacion de las librerias nos permitira manipular la informacion (numpy) y poder graficarla para ser mas facil la visualizacion del ejercicio (matplotlib)

```python
A = np.matrix([[10, 2.0], [2.0, 1.0]])
b = np.matrix([[2.0], [-8.0]]) 
c = 0.0
```

Aqui podemos ver que definimos tres variables que son la parte $A$ de nuestra ecuacion cuadratica, $B$ y $C$, cabe mencionar que $C$ esta definido como un valor constante que es 0

```python
def f(x, A, b, c):
    return float(0.5 * x.T * A * x - b.T * x + c)
```

Funcion en donde utilizaremos las varibles 

```python
def bowl(A, b, c):
    fig = plt.figure(figsize=(10,8))
    qf = fig.gca(projection='3d')
    size = 20
    x1 = list(np.linspace(-6, 6, size))
    x2 = list(np.linspace(-6, 6, size))
    x1, x2 = np.meshgrid(x1, x2)
    zs = np.zeros((size, size))
    for i in range(size):
        for j in range(size):
            x = np.matrix([[x1[i,j]], [x2[i,j]]])
            zs[i,j] = f(x, A, b, c)
    qf.plot_surface(x1, x2, zs, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0)
    fig.show()
    return x1, x2, zs
```

```python
x1, x2, zs = bowl(A, b, c)
```

```python
def contoursteps(x1, x2, zs, steps=None):
    fig = plt.figure(figsize=(6,6))
    cp = plt.contour(x1, x2, zs, 10)
    plt.clabel(cp, inline=1, fontsize=10)
    if steps is not None:
        steps = np.matrix(steps)
        plt.plot(steps[:,0], steps[:,1], '-o')
    fig.show()
```

```python
contoursteps(x1, x2, zs)
```
